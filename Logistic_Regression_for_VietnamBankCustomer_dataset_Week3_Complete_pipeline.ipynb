## 3. Xây mô hình Neural Network dùng thuật toán Logistic Regression



Ở 2 tuần trước, chúng ta đã được tìm hiểu qua về Forward Propagation và Backward Propagation. Bây giờ chúng ta sẽ xây dựng một Neural Network hoàn chỉnh từ 2 bước này.

Bạn hãy hoàn thành các phương trình sau, dựa vào những gì đã học ở các tuần trước.

```python
#Tạo hàm sigmoid
def sigmoid(z):
  ###BEGIN SOLUTION
  return 1/(1 + np.exp(-z))
###END SOLUTION
```

```python
#Tạo hàm ReLU
def ReLU(z):
  ###BEGIN SOLUTION
  return np.maximum(0,z)
  ###END SOLUTION
```

```python
#Tính đạo hàm của ReLU
def deriv_ReLU(z):
  ###BEGIN SOLUTION
  return (z > 0).astype(z.dtype)
  ###END SOLUTION
```

```python
# @title
def test_activation_function():
  assert sigmoid(0) == 0.5
  assert sigmoid(2) == 1/(1 + np.exp(-2))
  assert ReLU(0) == 0, "Hàm ReLU khởi tạo sai"
  assert ReLU(2) == 2, "Hàm ReLU khởi tạo sai"
  assert ReLU(-2) == 0, "Hàm ReLU khởi tạo sai"
  assert np.allclose(deriv_ReLU(np.random.uniform(high=0, size=(10, 13))),0), "Đạo hàm của ReLU khởi tạo sai"
  assert np.allclose(deriv_ReLU(np.random.uniform(low=1, high=100, size=(10, 13))),1), "Đạo hàm của ReLU khởi tạo sai"
  print("Chúc mừng bạn đã làm đúng!")
```

*Ảnh 1: Cấu trúc Neural Network*

Tiếp theo, ta cần khởi tạo tham số cho Neural Network. Để khởi tạo tham số, ta có những quy tắc sau đây:
- Đối với neural network 3 lớp như trên, giữa 2 lớp ta sẽ có một cặp tham số weight và bias
- Số hàng của tham số weight đầu tiên phải bằng với số neuron của lớp đầu tiên. Còn số cột của weight là số neuron ở lớp thứ 2. Vậy weight đầu tiên sẽ có dạng là (15, 128)
- Tham số bias luôn có số hàng là 1, và số cột là số neuron của lớp tiếp theo, Suy ra ta có vector dạng (1, 128)
- Tương cự với cặp tham số tiếp theo. Weight có số hàng là số neuron của lớp 2 và số cột là số neuron của lớp cuối cùng.

Gợi ý: dùng hàm randn của module numpy.random để khởi tạo tham số

```python
def initParam():
  ###BEGIN SOLUTION
  W1 = np.random.randn(15, 128)
  b1 = np.random.randn(1, 128)
  W2 = np.random.randn(128, 1)
  b2 = np.random.randn(1, 1)
  return W1, b1, W2, b2
  ###END SOLUTION
```

```python
# @title
def test_initParam():
  W1, b1, W2, b2 = initParam()

  # Check types
  assert isinstance(W1, np.ndarray), "W1 phải là vector dạng numpy"
  assert isinstance(b1, np.ndarray), "b1 phải là vector dạng numpy"
  assert isinstance(W2, np.ndarray), "W2 phải là vector dạng numpy"
  assert isinstance(b2, np.ndarray), "b2 phải là vector dạng numpy"

  ###HIDDEN TEST
  # Check shapes
  assert W1.shape == (15, 128), f"Vector W1 phải có các chiều là (13, 128), không phải {W1.shape}"
  assert b1.shape == (1, 128), f"Vector b1 phải có các chiều là (1, 128), không phải {b1.shape}"
  assert W2.shape == (128, 1), f"Vector W2 phải có các chiều là (128, 1), không phải {W2.shape}"
  assert b2.shape == (1, 1), f"Vector b2 phải có các chiều là (1, 1), không phải {b2.shape}"
  ### HIDDEN TEST

  print("Chúc mừng bạn đã làm đúng!")
```

```python
test_initParam()
```

Tiếp theo, bạn hãy ôn lại kiến thức tuần 1 và tạo hàm Forward Prop.

```python
def forward_prop(W1, b1, W2, b2, X):
  ###BEGIN SOLUTION
  Z1 = X.dot(W1) + b1
  A1 = ReLU(Z1)
  Z2 = A1.dot(W2) + b2
  A2 = sigmoid(Z2)
  return Z1, A1, Z2, A2
  ###END SOLUTION
```

```python
# @title
def test_fwd_prop():
  W1, b1, W2, b2 = initParam()
  X = np.random.randn(100, 13)
  Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)

  ###HIDDEN TESTS
  # Check values
  assert np.allclose(Z1, X.dot(W1) + b1), "Bạn đã khởi tạo Z1 sai"
  assert np.allclose(A1, ReLU(Z1)), "Bạn đã khởi tạo A1 sai"
  assert np.allclose(Z2, A1.dot(W2) + b2), "Bạn đã khởi tạo Z2 sai"
  assert np.allclose(A2, sigmoid(Z2)), "Bạn đã khởi tạo A2 sai"
  ###HIDDEN TESTS

  print("Chúc mừng bạn đã làm đúng!")
```

```python
test_fwd_prop()
```

Giờ bạn hãy ôn lại kiến thức tuần 2 để làm hàm Backward Prop

```python
def backward_prop(Z1, A1, Z2, A2, W2, X, Y):
  ###BEGIN SOLUTION
  m = Y.shape[0]
  #từ lớp đầu ra (output layer)
  dZ2 = A2 - Y
  dW2 = 1/m * np.dot(A1.T, dZ2)
  db2 = 1/m * np.sum(dZ2, 0)
  #từ các lớp ẩn (hidden layer)
  dZ1 =  np.dot(dZ2, W2.T) * deriv_ReLU(Z1)
  dW1 = 1/m * np.dot(X.T, dZ1)
  db1 = 1/m * np.sum(dZ1, 0)

  return dW1, db1, dW2, db2
  ###END SOLUTION
```

```python
# @title
def test_bwd_prop():
  W1, b1, W2, b2 = initParam()
  X = np.random.randn(100, 13)
  Y = np.random.randn(100, 1)

  Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)
  dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W2, X, Y)
  dZ2 = A2 - Y
  dZ1 =  np.dot(dZ2, W2.T) * deriv_ReLU(Z1)

  ### HIDDEN TEST
  assert np.allclose(dW1, 1/100 * X.T.dot(dZ1)), "dW1 hoặc dZ1 sai"
  assert np.allclose(db1, 1/100 * np.sum(dZ1, 0)), "db1 hoặc dZ1 sai"
  assert np.allclose(dW2, 1/100 * A1.T.dot(dZ2)), "dW2 hoặc dZ2 sai"
  assert np.allclose(db2, 1/100 * np.sum(dZ2, 0)), "db2 hoặc dZ2 sai"
  ### HIDDEN TEST

  print("Chúc mừng bạn đã làm đúng!")
```

```python
test_bwd_prop()
```

Đây là bước Gradient Descent đã được học ở tuần 2. Bạn hãy hoàn thành hàm này nhé

```python
def update_param(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha):
  ###BEGIN SOLUTION
  W1 -= alpha*dW1
  b1 -= alpha*db1
  W2 -= alpha*dW2
  b2 -= alpha*db2
  return W1, b1, W2, b2
  ###END SOLUTION
```

```python
# @title
def test_update_param():
  W1, b1, W2, b2 = initParam()
  X = np.random.randn(100, 13)
  Y = np.random.randn(100, 1)
  Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)
  dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W2, X, Y)

  ###HIDDEN TEST
  W1_exp = W1 - 0.1*dW1
  b1_exp = b1 - 0.1*db1
  W2_exp = W2 - 0.1*dW2
  b2_exp = b2 - 0.1*db2
  ###HIDDEN TEST

  W1_up, b1_up, W2_up, b2_up = update_param(W1, b1, W2, b2, dW1, db1, dW2, db2, 0.1)

  #print('1',W1_up, '\n 2', W1)

  assert np.allclose(W1_up, W1_exp), "W1 cập nhật chưa đúng"
  assert np.allclose(b1_up, b1_exp), "b1 cập nhật chưa đúng"
  assert np.allclose(W2_up, W2_exp), "W2 cập nhật chưa đúng"
  assert np.allclose(b2_up, b2_exp), "b2 cập nhật chưa đúng"

  print("Chúc mừng bạn đã làm đúng!")
```

Sau khi định nghĩa xong các hàm, ta sẽ bắt đầu train model. Khi train, các trọng số w và biến thiên b của model được cập nhật liên tục, vây nên ta sẽ cho các hàm bên trên chạy trong một vòng lặp.

Mục tiêu của training là giảm loss xuống ít nhất có thể. Từ đó tối ưu hoá khả năng dự đoán của model. Loss càng ít thì khả năng dự đoán của model càng tốt. Loss được định nghĩa như sau:
\begin{equation}Loss = Z_i = \frac{1}{m}\Sigma_{i=1}^{m}(\hat{y} - y)^2 \end{equation}

Trong đó m là số lượng mẫu trong train dataset, y là ground-truth còn $\hat{y}$
là kết quả dự đoán cuối cùng

Ngoài ra, ta có một đại lượng khác để đánh giá khả năng của model, gọi là accuracy.
\begin{equation}Accuracy = \frac{Số dự đoán đúng}{Tổng số các dự đoán}\end{equation}

Bạn hãy hoàn thành các hàm get_accurary và hàm train nhé.

```python
def get_accuracy(y_hat, y):
  predictions = (y_hat >= 0.5).astype(int)
  return np.sum(predictions == y) / len(predictions)

def train(X, Y, epoch, iters, alpha=1e-3):
  ### BEGIN SOLUTION
  W1, b1, W2, b2 = initParam()    # Dùng hàm khởi tạo tham số
  for i in tqdm_notebook(range(epoch), desc='Progress'):
    print("Epoch ", i+1)
    for j in tqdm_notebook(range(iters), desc='Train step'):
      Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)    # Dùng hàm Forward Prop
      dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W2, X, Y)    # Dùng hàm Backward Prop
      W1, b1, W2, b2 = update_param(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)  # Dùng hàm Gradient Descent
      if j % 50 == 0:
        print('Iter: ', j, " Loss: ", np.mean((A2-Y) ** 2))   # Tính loss dựa vào công thức trên. Gợi ý: dùng np.mean()
        print('Accuracy: ', get_accuracy(A2, Y))
    ### END SOLUTION

  return W1, b1, W2, b2
```

Bây giờ ta bắt đầu train model.

```python
W1, b1, W2, b2 = train(X_train, y_train, 50, 500)
```

# 4. Thử kết quả dự đoán

Sau đây ta sẽ thử cho model của chúng ta dự đoán trên tập X_val. Model chưa được train trên tập này, nên việc đánh giá model dựa trên một tập data mới giúp đánh giá khách quan hơn

```python
def make_predict(W1, b1, W2, b2, X):
  _, _, _, A2 = forward_prop(W1, b1, W2, b2, X)
  predictions = (A2 >= 0.5).astype(int)
  return predictions

def test_predict(W1, b1, W2, b2, index):
  prediction = make_predict(W1, b1, W2, b2, X_val[index])
  result = "Có gửi tiền" if prediction == 1 else "Không gửi tiền"
  print("Khách hàng số:", index)
  print("Thông tin chung:", X_val[index])
  print("Kết quả dự đoán:", result)
  label = y_val[index][0]
  print("Nhãn (sự thật):", label)
```
